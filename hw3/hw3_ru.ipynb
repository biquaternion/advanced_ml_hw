```python
unigrams_l = 'абвгдежзийклмнопрстуфхцчшщъыьэюя '
bigrams_l = [s1 + s2 for s1, s2 in product(unigrams_l, unigrams_l)]
# trigrams_l = [s1 + s2 for s1, s2 in product(unigrams_l, bigrams_l)]
# print(unigrams_l, len(unigrams_l))
# print(bigrams_l, len(bigrams_l))
# print(trigrams_l, len(trigrams_l))
```

```python
def read_text(path: str):
    with open(path, 'r') as f:
        return f.read()
```

```python
def filter_by_char(text: str, symbol_set):
    return ''.join([i for i in text if i in symbol_set])
```

```python
train_text = read_text('data/AnnaKarenina.txt')
train_text = filter_by_char(train_text.lower(), symbol_set=unigrams_l)
test_text = read_text('data/WarAndPeace.txt')
test_text = filter_by_char(test_text.lower(), symbol_set=unigrams_l)
```

```python
def train_ngrams(text: str, ngrams) -> dict:
#     set_trace()
    n = len(ngrams[0])
    c_vec = CountVectorizer(ngram_range=(n, n), analyzer='char')
    fit = c_vec.fit_transform([text])
    freqs = fit.toarray()[0] / len(text)
    voc = c_vec.vocabulary_
    return {ngram: freqs[voc[ngram]] if ngram in voc else 0 for ngram in ngrams}
```

```python
def mess_up(text: str):
#     set_trace()
    shuffled = [i for i in unigrams_l]
    np.random.shuffle(shuffled)
    lut = {k: v for k, v in zip(unigrams_l, shuffled)}
    return ''.join([lut[i] if i in lut else i for i in text]), lut
```

```python
messed_text, lut = mess_up(test_text)
```

```python
print(messed_text[:500])#, lut
```

```python
def decode(text: str, lut: dict, freqs=None):
    n = len(list(lut.keys())[0])
    lut_inv = {v: k for k, v in lut.items()}
    if n == 1:
        return ''.join([lut_inv[i] if i in lut_inv else i for i in text])
    else: # n == 2
        if freqs is None:  # naive implementation
            result = [lut_inv[text[i:i+n]] for i in range(0, len(text) - n, n)]
            return ''.join(result)
        
        ngrams = [] * n
        ngrams_idx = 0
        result = ''
        for i in range(len(text)):
            ngram = lut_inv[text[i:i+n]]
            ngrams[ngrams_idx] += [_ for _ in ngram]
            for i_ in range(n):
                if len[ngrams[i_]] <= i:
                    ngrams += '*'
            result += ngrams[ngrams]
            ngrams_idx += 1
            ngrams %= n
        
```

```python
def calc_lut(train_text, test_text, ngrams):
    n = len(ngrams[0])
    train_n = train_ngrams(text=train_text, ngrams=ngrams)
    test_n = train_ngrams(text=test_text, ngrams=ngrams)
    train_ = sorted([(v, k) for k, v in train_n.items()])
    test_ = sorted([(v, k) for k, v in test_n.items()])
    zipped = [_ for _ in zip(train_, test_)]
#     print(len(zipped))
    lut = {trn[1]: tst[1] for trn, tst in zipped}
    return lut
```

```python
lut_un = calc_lut(train_text=train_text, test_text=messed_text, ngrams=unigrams_l)
print(decode(text=messed_text[:500], lut=lut_un))
# print(lut_un)
```

```python
def get_probs_table(freqs):
    probs = defaultdict(dict)
    for k in freqs:
        f, s = k
        probs[f][s] = freqs[k]
    return probs
```

Предлагается подбирать случайным образом перестановки символов, оценивать score-правдоподобие по формуле из ячейки ниже: $$ \sum_{i} log(P(b|x_i,x_{i+1}) + P(u|x_i)) $$ (идея взята [отсюда](https://warsquid.com/blog/2019-07/solving-substitution-ciphers-with-markov-chains-in-python/)). Далее, если после итерации score увеличивается, то сохранить перестановку, иначе вернуться к прежней.

```python
def get_likelihood(text, unigram_probs, prob_table):
    total = 0
    for ch1, ch2 in zip(text[:-1], text[1:]):
#         set_trace()
        total += np.log(bigram_probs[ch1].get(ch2, 1e-2) + unigram_probs.get(ch1, 1))
    return total
```

```python
def decrypt(probe_text:str,
            max_iter=10000,
            from_scratch=False,
            verbose=True):
    swapped = defaultdict(bool)
    lut_c = lut_un
    if from_scratch:
        probe_text, lut_c = mess_up(probe_text)
    key = lut_c
    best_decoded = decode(text=probe_text, lut=key)
    best_likelihood = get_likelihood(best_decoded, 
                                     unigram_probs, 
                                     bigram_probs)
    for it in tqdm(range(max_iter)):
        p = np.array([1.0 - float(v) for k, v in unigram_probs.items()])
#         p = p[:-1]
        p /= np.sum(p)
#         set_trace()
        idx1, idx2 = np.random.choice(len(unigrams_l), size=2)#, p=p)
        c1, c2 = unigrams_l[idx1], unigrams_l[idx2]
#         if swapped[(c1, c2)]:
#             continue
        key[c1], key[c2] = key[c2], key[c1]
        decoded = decode(probe_text, key)
        likelihood = get_likelihood(decoded, unigram_probs, bigram_probs)
        if likelihood > best_likelihood:
            best_likelihood = likelihood
            best_decoded = decoded
        else:
            key[c1], key[c2] = key[c2], key[c1]
#         swapped[c1, c2] = True
        if verbose:
            if it % 100 == 0:
                print(it)
                print(decoded[:500])
                print(likelihood)
    return best_decoded
```
